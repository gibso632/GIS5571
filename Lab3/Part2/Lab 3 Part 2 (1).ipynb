```python
field_list = ['X','Y']
calc_list = ['Longitude','Latitude']
for field in field_list:
    arcpy.management.AddField(
        in_table = 'NDAWN_30_days',
        field_name = field,
        field_type = 'DOUBLE'
    )
    arcpy.management.CalculateField(
        in_table = 'NDAWN_30_days',
        field = field,
        expression = "!" + calc_list[field_list.index(field)] + "!"
    )
```

```python
field_list = ['Max','Min','Avg']
calc_list = ['Max_Temp','Min_Temp','Avg_Temp']
for field in field_list:
    arcpy.management.AddField(
        in_table = 'NDAWN_30_Days',
        field_name = field,
        field_type = 'DOUBLE'
    )
    arcpy.management.CalculateField(
        in_table = 'NDAWN_30_Days',
        field = field,
        expression = "!" + calc_list[field_list.index(field)] + "!"
    )
```

```python
arcpy.management.XYTableToPoint(
    in_table = 'NDAWN_30_days',
    out_feature_class = 'NDAWN30Days',
    x_field = 'X',
    y_field = 'Y'
)
```

```python
NDAWN_AVG_daily_temp = arcpy.management.Dissolve(
    in_features = 'NDAWN30Days',
    out_feature_class = r'C:\Users\15612\Documents\ArcGIS\Projects\GIS5571 Lab 3 Part 2\GIS5571 Lab 3 Part 2.gdb\NDAWN_AVG_daily_temp',
    dissolve_field = 'Station_Na',
    statistics_fields = [['Avg','MEAN'],['Max','MEAN'],['Min','MEAN']]
)
```

```python
krigOrdinary = arcpy.sa.KrigingModelOrdinary()

IDW_AVG = arcpy.sa.Idw(
    in_point_features = 'NDAWN_AVG_daily_temp',
    z_field = 'MEAN_Avg'
)

IDW_High = arcpy.sa.Idw(
    in_point_features = 'NDAWN_AVG_daily_temp',
    z_field = 'MEAN_Max'
)

IDW_Low = arcpy.sa.Idw(
    in_point_features = 'NDAWN_AVG_daily_temp',
    z_field = 'MEAN_Min'
)

Spline_AVG = arcpy.sa.Spline(
    in_point_features = 'NDAWN_AVG_daily_temp',
    z_field = 'MEAN_Avg'
)

Spline_Max = arcpy.sa.Spline(
    in_point_features = 'NDAWN_AVG_daily_temp',
    z_field = 'MEAN_Max'
)

Spline_Min = arcpy.sa.Spline(
    in_point_features = 'NDAWN_AVG_daily_temp',
    z_field = 'MEAN_Min'
)

Kriging_AVG = arcpy.sa.Kriging(
    'NDAWN_AVG_daily_temp',
    'MEAN_Avg',
    krigOrdinary
)

Kriging_Max = arcpy.sa.Kriging(
    'NDAWN_AVG_daily_temp',
    'MEAN_Max',
    krigOrdinary
)

Kriging_Min = arcpy.sa.Kriging(
    'NDAWN_AVG_daily_temp',
    'MEAN_Min',
    krigOrdinary
)
```
